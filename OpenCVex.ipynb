```python
import matplotlib.pyplot as plt
import cv2

image=cv2.imread("/content/sample_data/image/images.jpg")
img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#display the image using matplotlib library
plt.imshow(img_rgb)
plt.axis("off")
plt.title("CV2 image file")
plt.show()
```
