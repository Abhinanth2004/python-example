```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")
plt.rcParams["figure.figsize"] = (10,6)
sns.set_style("whitegrid")
pd.set_option('display.float_format', lambda x: '%.3f' % x)
# Set it None to display all rows in the dataframe
# pd.set_option('display.max_rows', None)
# Set it to None to display all columns in the dataframe
pd.set_option('display.max_columns', None)
import io
import plotly.express as px

train = pd.read_csv('/content/sample_data/rnn_train/DailyDelhiClimateTrain.csv',
                    parse_dates=['date'], index_col="date")
test = pd.read_csv('/content/sample_data/rnn_test/DailyDelhiClimateTest.csv',
                   parse_dates=['date'], index_col="date")
train.head()
test.head()
train.describe().style.background_gradient(cmap = 'Blues')

test.describe().style.background_gradient(cmap = 'Blues')
test.info()
train.info()
train.shape
test.shape
df = pd.concat([train,test])
df.head()

df.info()
df.describe().style.background_gradient(cmap = 'Blues')

df.isnull().sum()

df.shape

df["year"] = df.index.year
df['month'] = df.index.month

df.sample()

import plotly.graph_objects as go

# Mean Temperature

fig_temp = go.Figure()
fig_temp.add_trace(go.Scatter(x=df.index,
                              y=df['meantemp'],
                              mode='lines',
                              line=dict(color='yellow')))
fig_temp.update_layout(title="Mean Temperature",
                       xaxis_title="Date",
                       yaxis_title="Value",
                       template='plotly_dark',
                       title_x=0.5)
fig_temp.show()
```
